```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/matches_with_stats.csv", parse_dates=["tourney_date"])

print("Initial shape:", df.shape)
df.head()
```

```python
drop_cols = [
    "winner_entry",
    "loser_seed",
    "loser_entry",
    "minutes",
    "loser_rank_points",
    "winner_rank_points",
    "loser_ht"
]

# Drop only if they exist
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

print("After dropping columns:", df.shape)
```

```python
# Drop rows with any NaN left
df_cleaned = df.dropna().reset_index(drop=True)

print("After dropping null rows:", df_cleaned.shape)

# Check nulls to confirm all gone
print("Remaining nulls per column:\n", df_cleaned.isnull().sum().sum())
```

```python
output_path = "data/matches_with_stats_cleaned.csv"
df_cleaned.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to {output_path}")
```
