```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/matches_with_stats.csv", parse_dates=["tourney_date"])

# Basic overview
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nInfo:")
print(df.info())

# first rows
df.head()
```

```python
# Null values per column
null_counts = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:\n", null_counts)

# Summary statistics numeric
df.describe().T
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df['minutes'].dropna().hist(bins=50)
plt.title("Distribution of Match Duration (minutes)")
plt.xlabel("Minutes")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10,5))
df['w_ace'].dropna().hist(bins=50)
plt.title("Distribution of Winner Aces")
plt.xlabel("Aces")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

#compare aces
plt.figure(figsize=(8,6))
sns.kdeplot(df['w_ace'], label="Winner Aces", fill=True, alpha=0.5)
sns.kdeplot(df['l_ace'], label="Loser Aces", fill=True, alpha=0.5)
plt.legend()
plt.title("Winner vs Loser Aces Distribution")
plt.show()


plt.figure(figsize=(8,6))
sns.boxplot(data=df[['winner_rank','loser_rank']])
plt.title("Winner vs Loser Rank")
plt.show()
```

```python

surface_counts = df['surface'].value_counts()
print(surface_counts)


avg_minutes_surface = df.groupby('surface')['minutes'].mean().sort_values()
print(avg_minutes_surface)


plt.figure(figsize=(8,6))
sns.barplot(x=avg_minutes_surface.index, y=avg_minutes_surface.values)
plt.title("Average Match Duration by Surface")
plt.ylabel("Minutes")
plt.xlabel("Surface")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlations for numeric columns only
corr = df.corr(numeric_only=True)

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Correlation Heatmap of Match Statistics")
plt.show()
```

```python
import numpy as np

eps = 1e-9
df["w_first_serve_pct"] = df["w_1stIn"] / (df["w_svpt"] + eps)
df["w_1st_won_pct"]    = df["w_1stWon"] / (df["w_1stIn"] + eps)
df["w_2nd_won_pct"]    = df["w_2ndWon"] / (df["w_svpt"] - df["w_1stIn"] + eps)
df["w_bp_save_pct"]    = df["w_bpSaved"] / (df["w_bpFaced"] + eps)

serve_cols = [
    "w_ace","w_df","w_svpt","w_1stIn","w_1stWon","w_2ndWon","w_SvGms",
    "w_bpSaved","w_bpFaced",
    "w_first_serve_pct","w_1st_won_pct","w_2nd_won_pct","w_bp_save_pct"
]
serve_cols = [c for c in serve_cols if c in df.columns]

serve_df = df[serve_cols].replace([np.inf, -np.inf], np.nan).dropna(how="all")

corr = serve_df.corr(numeric_only=True)

plt.figure(figsize=(12,9))
sns.heatmap(
    corr,
    annot=True,          # âœ… show numbers
    fmt=".2f",           # format to 2 decimal places
    cmap="coolwarm",
    center=0,
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Winner Serve Metrics â€” Correlation Heatmap", fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="w_ace", y="minutes", data=df, alpha=0.3, label="Winner Aces")
sns.scatterplot(x="l_ace", y="minutes", data=df, alpha=0.3, label="Loser Aces")
plt.title("Aces vs Match Duration")
plt.xlabel("Aces")
plt.ylabel("Match Duration (minutes)")
plt.legend()
plt.show()
```

```python
# Winner IOC = country code
top_countries = df['winner_ioc'].value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 15 Winner Countries")
plt.ylabel("Number of Wins")
plt.xlabel("Country")
plt.show()
```

```python
df['rank_diff'] = df['loser_rank'] - df['winner_rank']

plt.figure(figsize=(10,6))
sns.histplot(df['rank_diff'], bins=50, kde=True)
plt.axvline(0, color='red', linestyle='--')
plt.title("Distribution of Rank Difference (Loser Rank - Winner Rank)")
plt.xlabel("Rank Difference")
plt.ylabel("Frequency")
plt.show()
```

```python
avg_minutes_round = df.groupby("round")['minutes'].mean().sort_values()

plt.figure(figsize=(12,6))
sns.barplot(x=avg_minutes_round.index, y=avg_minutes_round.values)
plt.title("Average Match Duration by Round")
plt.ylabel("Minutes")
plt.xlabel("Round")
plt.show()
```

```python
# Approximate: winner first serve percentage = w_1stIn / w_svpt
df['w_first_serve_pct'] = df['w_1stIn'] / df['w_svpt']
df['l_first_serve_pct'] = df['l_1stIn'] / df['l_svpt']

plt.figure(figsize=(10,6))
sns.kdeplot(df['w_first_serve_pct'], fill=True, alpha=0.5, label="Winners")
sns.kdeplot(df['l_first_serve_pct'], fill=True, alpha=0.5, label="Losers")
plt.legend()
plt.title("First Serve Percentage: Winners vs Losers")
plt.xlabel("First Serve %")
plt.show()
```

```python
df['w_bp_save_pct'] = df['w_bpSaved'] / df['w_bpFaced']
df['l_bp_save_pct'] = df['l_bpSaved'] / df['l_bpFaced']

plt.figure(figsize=(10,6))
sns.kdeplot(df['w_bp_save_pct'].dropna(), fill=True, alpha=0.5, label="Winners")
sns.kdeplot(df['l_bp_save_pct'].dropna(), fill=True, alpha=0.5, label="Losers")
plt.legend()
plt.title("Break Point Save %: Winners vs Losers")
plt.xlabel("Break Point Save %")
plt.show()
```
